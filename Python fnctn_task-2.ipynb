6.Write a python function, which will find all such numbers between 1000 and 3000 (both included) such that each digit of the number is an even number. Return the results as a list

```python
def find_even_digit_numbers():
    result = []

    for num in range(1000, 3001):
        all_even_digits = all(int(digit) % 2 == 0 for digit in str(num))
        if all_even_digits:
            result.append(num)
    return result

even_digit_numbers = find_even_digit_numbers()
print(even_digit_numbers)
```

7.Write a python function that accepts a sentence and calculate and return the number of letters and digits.
Suppose the following input is supplied to the program:
hello world! 123
Then, the output should be:
LETTERS 10
DIGITS 3


```python
def letter_digit_number(x):
  letter_count=0
  digit_count=0
  for i in x:
    if i.isalpha():
      letter_count+=1
    elif i.isdigit():
      digit_count+=1
  return letter_count,digit_count
a="hello wolrd! 123"
letter_count,digit_count=letter_digit_number(a)
print(f"letter:{letter_count} digit:{digit_count}")

```

8.Write a Python program to convert all the characters into uppercase and lowercase and eliminate duplicate letters from a given sequence. Use the map() function

---



```python
def upper_lower_duplicate(sentence):
  upper_case=list(map(lambda x:x.upper(),sentence))
  lower_case=list(map(lambda x:x.lower(),sentence))
  duplicate=list(set(sentence))
  return upper_case,lower_case,duplicate
sentence="Hello,Wolrd"
upper_case,lower_case,duplicate=upper_lower_duplicate(sentence)
print(upper_case)
print(lower_case)
print(duplicate)
```

15.Write a Python program to sort a given mixed list of integers and strings using lambda. Numbers must be sorted before strings.

Original list:

[19, 'red', 12, 'green', 'blue', 10, 'white', 'green', 1]

Sort the said mixed list of integers and strings:

[1, 10, 12, 19, 'blue', 'green', 'green', 'red', 'white']



```python
l=[19, 'red', 12, 'green', 'blue', 10, 'white', 'green', 1]
def filter_list(l):

  a=[]
  b=[]
  for element in l:
    if isinstance(element,int):
      a.append(element)
    else:
      b.append(element)
  a.sort()
  b.sort()
  return a+b
result=filter_list(l)
print(result)
```

16.Write a Python program to count the occurrences of items in a given list using lambda.

Original list:

[3, 4, 5, 8, 0, 3, 8, 5, 0, 3, 1, 5, 2, 3, 4, 2]

Count the occurrences of the items in the said list:

{3: 4, 4: 2, 5: 3, 8: 2, 0: 2, 1: 1, 2: 2}


```python
l=[3, 4, 5, 8, 0, 3, 8, 5, 0, 3, 1, 5, 2, 3, 4, 2]
result=dict(map((lambda x:(x,l.count(x))),l))
print(result)
```
